 # SET PSD calculation and FOOOF PARAMETERS

```python
f_min = 1
f_max = 45
t     = 15
fs    = 200
SETTINGS_W  = { 'method' : 'welch', 'average' : 'mean', 'fmin' :f_min , 'fmax':f_max, 'n_fft': fs*t, 'n_overlap': fs*t*0.5}

SETTINGS_F1={'max_n_peaks':8, 'aperiodic_mode':'knee'}
SETTINGS_F2={'max_n_peaks':8, 'aperiodic_mode':'fixed'}
```

# Set the vectors to store the results

```python
# Set data path
data_path = Path('/home/b1044271/Columbia/iEEG')
# Set result path
result_path = Path('/home/b1044271/Columbia/Results/iEEG')
# Set dataset
dataset = 'Wakefulness_AllRegions'
# Get list of files, which correspond to regions
regions = np.sort(os.listdir(data_path / dataset))
```

```python
W_Kn  = np.zeros([len(regions),200])*np.nan
W_Exp = np.zeros([len(regions),200])*np.nan
W_Exp2 = np.zeros([len(regions),200])*np.nan
W_R2K = np.zeros([len(regions),200])*np.nan
W_R2N = np.zeros([len(regions),200])*np.nan
W_EK = np.zeros([len(regions),200])*np.nan
W_EN = np.zeros([len(regions),200])*np.nan
W_PK = np.zeros([len(regions),200])*np.nan
W_PN = np.zeros([len(regions),200])*np.nan

N2_Kn  = np.zeros([len(regions),200])*np.nan
N2_Exp = np.zeros([len(regions),200])*np.nan
N2_Exp2 = np.zeros([len(regions),200])*np.nan
N2_R2K = np.zeros([len(regions),200])*np.nan
N2_R2N = np.zeros([len(regions),200])*np.nan
N2_EK = np.zeros([len(regions),200])*np.nan
N2_EN = np.zeros([len(regions),200])*np.nan
N2_PK = np.zeros([len(regions),200])*np.nan
N2_PN = np.zeros([len(regions),200])*np.nan

N3_Kn  = np.zeros([len(regions),200])*np.nan
N3_Exp = np.zeros([len(regions),200])*np.nan
N3_Exp2 = np.zeros([len(regions),200])*np.nan
N3_R2K = np.zeros([len(regions),200])*np.nan
N3_R2N = np.zeros([len(regions),200])*np.nan
N3_EK = np.zeros([len(regions),200])*np.nan
N3_EN = np.zeros([len(regions),200])*np.nan
N3_PK = np.zeros([len(regions),200])*np.nan
N3_PN = np.zeros([len(regions),200])*np.nan

R_Kn  = np.zeros([len(regions),200])*np.nan
R_Exp = np.zeros([len(regions),200])*np.nan
R_Exp2 = np.zeros([len(regions),200])*np.nan
R_R2K = np.zeros([len(regions),200])*np.nan
R_R2N = np.zeros([len(regions),200])*np.nan
R_EK = np.zeros([len(regions),200])*np.nan
R_EN = np.zeros([len(regions),200])*np.nan
R_PK = np.zeros([len(regions),200])*np.nan
R_PN = np.zeros([len(regions),200])*np.nan
```

# WAKEFULNESS

```python
PSDs_W =  np.zeros([len(regions),661])*np.nan

i =0 
for C, V in enumerate(regions):
        
    EEG = mne.io.read_raw_edf(data_path / dataset / V)
    EEG_psd = EEG.compute_psd(**SETTINGS_W)
    
    PSDs_W[i,0:661] = np.mean(EEG_psd._data,0)
    
    
    fm1 = FOOOFGroup(**SETTINGS_F1)
    fm1.fit(EEG_psd._freqs, EEG_psd._data, [EEG_psd._freqs[0] , EEG_psd._freqs[-1]])
    
    fm2 = FOOOFGroup(**SETTINGS_F2)
    fm2.fit(EEG_psd._freqs, EEG_psd._data, [EEG_psd._freqs[0] , EEG_psd._freqs[-1]])
    
    # parameters exctract
    W_Kn[i,0:len(EEG_psd._data)]  = fm1.get_params('aperiodic_params','knee')
    W_Exp[i,0:len(EEG_psd._data)] = fm1.get_params('aperiodic_params','exponent')
    
    W_Exp2[i,0:len(EEG_psd._data)] = fm2.get_params('aperiodic_params','exponent')
   
    W_R2K[i,0:len(EEG_psd._data)] = fm1.get_params('r_squared')
    W_R2N[i,0:len(EEG_psd._data)] = fm2.get_params('r_squared')

    W_EK[i,0:len(EEG_psd._data)] = fm1.get_params('error')
    W_EN[i,0:len(EEG_psd._data)] = fm2.get_params('error')

    # Extract the number of peaks for each epoch
    for epoch_idx in range(len(EEG_psd._data)):
        peak_params1 = fm1.get_params('peak_params')
        num_peaks1 = sum(peak[-1] == epoch_idx for peak in peak_params1)  # Count peaks for each epoch
        W_PK[i, epoch_idx] = num_peaks1
        peak_params2 = fm2.get_params('peak_params')
        num_peaks2 = sum(peak[-1] == epoch_idx for peak in peak_params2)  # Count peaks for each epoch
        W_PN[i, epoch_idx] = num_peaks2 

        #empty vectors
        num_peaks1 = []
        num_peaks2 = []
        peak_params1 = []
        peak_params2 = []
        
    i = i+1
```

# Calculate BIC

```python
# Import Custom functions for BIC calculation
from utils import calculate_bic

# EMPTY VECTORS
BIC_NK_W = np.zeros(len(W_R2N))*np.nan
BIC_K_W = np.zeros(len(W_R2K))*np.nan
BIC_NK_N2 = np.zeros(len(N2_R2N))*np.nan
BIC_K_N2 = np.zeros(len(N2_R2K))*np.nan
BIC_NK_N3 = np.zeros(len(N3_R2N))*np.nan
BIC_K_N3 = np.zeros(len(N3_R2K))*np.nan
BIC_NK_R = np.zeros(len(R_R2N))*np.nan
BIC_K_R = np.zeros(len(R_R2K))*np.nan

# set parameters
num_params_nk = 26  # I modelled 8 peaks *3 = 24 and + 2 aperiodic parameter (slope + offset)
num_params_k = num_params_nk + 1# added the knee to aperiodic parameters (8 peasks + 3 aperiodic params)

def equalize_trials(Xn, Xk):
    """
    Equalizes the number of trials between two arrays by truncating the larger one.
    """
    min_len = min(len(Xn), len(Xk))
    return Xn[:min_len], Xk[:min_len]

# BIC Per subject
for i in np.arange(len(W_EN)):

    Wnx, Wkx = equalize_trials(W_EN[i][~np.isnan(W_EN[i])]**2, W_EK[i][~np.isnan(W_EK[i])]**2)
    N2nx, N2kx = equalize_trials(N2_EN[i][~np.isnan(N2_EN[i])]**2, N2_EK[i][~np.isnan(N2_EK[i])]**2)
    N3nx, N3kx = equalize_trials(N3_EN[i][~np.isnan(N3_EN[i])]**2, N3_EK[i][~np.isnan(N3_EK[i])]**2)
    Rnx, Rkx = equalize_trials(R_EN[i][~np.isnan(R_EN[i])]**2, R_EK[i][~np.isnan(R_EK[i])]**2)


    BIC_NK_W[i] = np.mean(calculate_bic(len(Wnx), Wnx, num_params_nk))
    BIC_K_W[i] = np.mean(calculate_bic(len(Wkx), Wkx, num_params_k))
    
    BIC_NK_N2[i] = calculate_bic(len(N2nx), np.mean(N2nx), num_params_nk)
    BIC_K_N2[i] = calculate_bic(len(N2kx), np.mean(N2kx), num_params_k)
    
    BIC_NK_N3[i] = calculate_bic(len(N3nx), np.mean(N3nx), num_params_nk)
    BIC_K_N3[i] = calculate_bic(len(N3kx), np.mean(N3kx), num_params_k)
    
    BIC_NK_R[i] = calculate_bic(len(Rnx), np.mean(Rnx), num_params_nk)
    BIC_K_R[i] = calculate_bic(len(Rkx), np.mean(Rkx), num_params_k)

    Wnx=[]
    Wkx=[]
    N2nx=[]
    N2kx=[]
    N3nx=[]
    N3kx=[]
    Rnx=[]
    Rkx=[]
```
