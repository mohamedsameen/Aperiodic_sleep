# Now is there an oscillatory component to the ERP ?

```python
# Set paths
Data_path = '/home/b1044271/Columbia/Results/Time-resolved/Transitions/Cz/Stim_NoStim/NEW/AR/*DummyN_FOOOFed_MT.json'
Data_path2 = '/home/b1044271/Columbia/Results/Time-resolved/Transitions/Cz/Stim_NoStim/NEW/AR/*DummyR_FOOOFed_MT.json'
Data_path3 = '/home/b1044271/Columbia/Results/Time-resolved/Transitions/Cz/Stim_NoStim/NEW/AR/*StimsN_FOOOFed_MT.json'
Data_path4 = '/home/b1044271/Columbia/Results/Time-resolved/Transitions/Cz/Stim_NoStim/NEW/AR/*StimsR_FOOOFed_MT.json'

    
subj_files = np.sort(glob.glob(Data_path))
O_DMN = np.ones([17,1250]) * np.nan

i=0
ii=0
for x in subj_files:
    fm1 = FOOOFGroup()
    fm1.load(x)
    peaks_NKC = fm1.get_params('peak_params', 'PW') # get oscillatory info

    Pwr_1 = peaks_NKC[:,0]
    L1 = peaks_NKC[:,1] # get datapoints (info on time points)
    L2=np.unique(L1) # get all datapoints that have oscilltory info

    # Now get all datapoints that are missing, i.e. have no oscillatory values
    res = np.in1d(np.arange(0,float(640),1), L2) 
    empty = [i for i, val in enumerate(res) if not val] # get the indices of these datapoints

    # Group oscillatory power values for the same datapoint (index) together
    R1 = { n:rep[n] for rep in [{}] for i,n in enumerate(L1) 
              if rep.setdefault(n,[]).append(i) or len(rep[n])==1 }

    # Now insert those nans in the dictionary at correct position
    items = list(R1.items())
    for z in empty:
        items.insert(z, (float(z), np.nan))

    R2 = dict(items)
    
    # fill final vector with mean of oscillatory activity at each timepoint
    for key, value in R2.items():
        
        if np.size(value) == 1: # sometimes there is no oscillations at this time point, then 
                if np.isnan(value):
                    continue
        
        if np.size(value) > 1:
            if any(np.isnan(value)):
                value = value[~np.isnan(value)]  
                
        O_DMN[ii,int(key)]=round(np.nanmean(Pwr_1[value]),2) # this is the final vector
    
    # move on to the next cell
    ii=ii+1
    L1=[]
    R2=[]
    R1=[]
    L2=[]
   
    #########################################################################

# REM
subj_files = np.sort(glob.glob(Data_path2))
O_DMR = np.ones([17,1250]) * np.nan

i=0
ii=0
for x in subj_files:
    fm1 = FOOOFGroup()
    fm1.load(x)
    peaks_NKC = fm1.get_params('peak_params', 'PW') # get oscillatory info

    Pwr_1 = peaks_NKC[:,0]
    L1 = peaks_NKC[:,1] # get datapoints (info on time points)
    L2=np.unique(L1) # get all datapoints that have oscilltory info

    # Now get all datapoints that are missing, i.e. have no oscillatory values
    res = np.in1d(np.arange(0,float(640),1), L2) 
    empty = [i for i, val in enumerate(res) if not val] # get the indices of these datapoints

    # Group oscillatory power values for the same datapoint (index) together
    R1 = { n:rep[n] for rep in [{}] for i,n in enumerate(L1) 
              if rep.setdefault(n,[]).append(i) or len(rep[n])==1 }

    # Now insert those nans in the dictionary at correct position
    items = list(R1.items())
    for z in empty:
        items.insert(z, (float(z), np.nan))

    R2 = dict(items)
    
    # fill final vector with mean of oscillatory activity at each timepoint
    for key, value in R2.items():
        
        if np.size(value) == 1: # sometimes there is no oscillations at this time point, then 
                if np.isnan(value):
                    continue
        
        if np.size(value) > 1:
            if any(np.isnan(value)):
                value = value[~np.isnan(value)]  
                
        O_DMR[ii,int(key)]=round(np.nanmean(Pwr_1[value]),2) # this is the final vector
    
    # move on to the next cell
    ii=ii+1
    L1=[]
    R2=[]
    R1=[]
    L2=[]
```

```python
def create_time_frequency_matrix(timepoints, frequencies, powers):
    # Step 1: Round frequency values to the lowest integer
    rounded_frequencies = np.floor(frequencies).astype(int)

    # Step 2: Identify unique time points and frequencies
    unique_timepoints = np.unique(timepoints)
    unique_frequencies = np.unique(rounded_frequencies)

    # Step 3: Initialize the matrix to store mean power values
    time_frequency_matrix = np.zeros((len(unique_frequencies), len(unique_timepoints)))

    # Step 4: Calculate the mean power for each frequency and time point
    for i, freq in enumerate(unique_frequencies):
        for j, time in enumerate(unique_timepoints):
            # Find indices where frequency and time point match
            indices = np.where((rounded_frequencies == freq) & (timepoints == time))[0]

            if len(indices) > 0:
                # Calculate mean power for these indices
                mean_power = np.mean(powers[indices])
                time_frequency_matrix[i, j] = mean_power
            else:
                # If no indices found, assign zero or np.nan
                time_frequency_matrix[i, j] = np.nan

    return time_frequency_matrix, unique_frequencies, unique_timepoints
```

```python
def insert_nans(arr, row_indices=None, col_indices=None):
    # Ensure row_indices and col_indices are lists and sort them
    row_indices = sorted(row_indices if row_indices is not None else [])
    col_indices = sorted(col_indices if col_indices is not None else [])
    
    # Insert NaN columns
    for col_idx in col_indices:
        if col_idx > arr.shape[1]:
            arr = np.hstack((arr, np.full((arr.shape[0], col_idx - arr.shape[1] + 1), np.nan)))
        arr = np.insert(arr, col_idx, np.nan, axis=1)
   
    # Insert NaN rows
    for row_idx in row_indices:
        if row_idx > arr.shape[0]:
            arr = np.vstack((arr, np.full((row_idx - arr.shape[0] + 1, arr.shape[1]), np.nan)))
        arr = np.insert(arr, row_idx, np.nan, axis=0)

    return arr

```
