## 2. Get aperiodic papers in brain vs sleep over the years

```python
start_year, end_year = 1924, 2024
years = list(range(start_year, end_year, 2))

year_counts=np.zeros([2,100])*np.nan

# Initialize counts object
counts = Counts()
terms_a = [['brain'], ['sleep']]  
terms_b = [['aperiodic exponent', 'spectral slope', 'spectral exponent', 'spectral parameterization', 'specparam', 'fooof', 'irasa', '1/f', '1/f-like', 'one-over f', 'aperiodic']]
inclusions = [['EEG','MEG','iEEG','electroencephalography', 'magnetoencephalography', 'intracranial electroencephalography', 'intracranial EEG'],['EEG','MEG','iEEG','electroencephalography', 'magnetoencephalography', 'intracranial electroencephalography', 'intracranial EEG']]
labels = ['brain', 'sleep']

# Set terms lists, indicating they are separate lists with the 'A' and 'B' labels
counts.add_terms(terms_a, dim='A')
counts.add_terms(terms_b, dim='B')
counts.add_terms(inclusions,  term_type='inclusions', dim='A')

# Collect co-occurrence data
i=0
for start_year in years:
    counts.run_collection( mindate=str(start_year) + '/01/01', maxdate=str(start_year+1) + '/12/31')
    year_counts[0:2,i] = np.squeeze(counts.counts)
    i=i+1
```

```python
# Define the range of years (2-year intervals)
start_year, end_year = 1924, 2024
years = list(range(start_year, end_year, 2))

# Initialize dynamic counts array
year_counts = []

# Initialize the LISC Counts object
counts = Counts()

# Define search terms
terms_a = [['brain'], ['sleep']]  
terms_b = [
    ['aperiodic exponent', 'aperiodic slope', 'spectral slope', 'spectral exponent', 'spectral parameterization', 'aperiodic',
     'specparam', 'fooof', 'irasa', '1/f', '1/f-like', 'one-over-f', 'non-oscillatory brain activity', 'scale-free']
]
inclusions = [
    ['EEG', 'MEG', 'iEEG', 'electroencephalography', 'magnetoencephalography', 
     'intracranial electroencephalography', 'intracranial EEG'],
    ['EEG', 'MEG', 'iEEG', 'electroencephalography', 'magnetoencephalography', 
     'intracranial electroencephalography', 'intracranial EEG']
]
labels = ['brain', 'sleep']

# Add terms to the Counts object
counts.add_terms(terms_a, dim='A')
counts.add_terms(terms_b, dim='B')
counts.add_terms(inclusions, term_type='inclusions', dim='A')

# Iterate over 2-year intervals and collect counts
for i, start in enumerate(years):
    # Collect co-occurrence data for the current interval
    counts.run_collection(mindate=f"{start}/01/01", maxdate=f"{start+1}/12/31")
    
    # Extract counts for the current interval
    current_counts = np.squeeze(counts.counts)
    
    # Ensure the counts are correctly formatted
    if current_counts.size == 2:  # Expected size based on terms
        year_counts.append(current_counts)
    else:
        print(f"Warning: Unexpected count size for years {start}-{start+1}")

# Convert counts list to an array
year_counts = np.array(year_counts).T

# Print resulting counts for verification
print("Year Counts Array Shape:", year_counts.shape)
print("Year Counts Array:", year_counts)
```

```python
# Data
freq_ranges = ['2-20', '1-45', '2-48', '30-45', '1-20', '0.5-12', '25-45', '0.5-35', '3-30', '1-30', '0.5-20']
counts = [1, 2, 3, 3, 1, 1, 1, 1, 1, 1, 1]

# Process frequency ranges
low_freqs = []
high_freqs = []

for fr in freq_ranges:
    low, high = map(float, fr.split('-'))
    low_freqs.append(low)
    high_freqs.append(high)

# Create a heatmap matrix
low_unique = sorted(set(low_freqs))
high_unique = sorted(set(high_freqs))
heatmap_matrix = np.zeros((len(high_unique), len(low_unique)))

for i, (low, high, count) in enumerate(zip(low_freqs, high_freqs, counts)):
    low_idx = low_unique.index(low)
    high_idx = high_unique.index(high)
    heatmap_matrix[high_idx, low_idx] += count

# Create the heatmap with masking for zero values
masked_matrix = np.ma.masked_where(heatmap_matrix == 0, heatmap_matrix)

fig, ax = plt.subplots(figsize=(5, 4))
sns.heatmap(masked_matrix, annot=True, fmt="g", cmap="Greys", cbar_kws={'label': '# of articles'},
            xticklabels=[int(x) if x.is_integer() else int(round(x)) for x in low_unique],
    yticklabels=[int(x) if x.is_integer() else int(round(x)) for x in high_unique],vmin=0,
            vmax=4,
            mask=(heatmap_matrix == 0), ax=ax)

# Customize the plot
ax.set_xlabel('lower frequency limit (Hz)', fontsize=14)
ax.set_ylabel('higher frequency limit (Hz)', fontsize=14)
#ax.set_title('Literature Frequency Ranges Heatmap (No Zero Voxels)', fontsize=18)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Flip y-axis
ax.invert_yaxis()

plt.savefig("/home/b1044271/Columbia/Plots/NOV24/Literature_freqs_heat.svg", dpi = 600)
```
