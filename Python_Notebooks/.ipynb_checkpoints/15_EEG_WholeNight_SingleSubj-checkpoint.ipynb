# Sleep stages

```python
# get all Sleep staging files
staging_files = '/home/b1044271/EEGsleep/SleepStaging/*.txt'
stage_files = glob.glob(staging_files)
```

```python
# ADJUST Sampling rate of sleep stages
sfiles = np.zeros([17,50000])*np.nan
i=0
for X in stage_files:
    x  = resample_stages(X, 2)
    sfiles[i,0:len(x)] = x
    i=i+1
```

```python
SS = sfiles[1] 
SS[SS == 3] = 4
SS[SS == 2] = 3
SS[SS == 1] = 2
SS[SS == 5] = 1
```

```python
fig, ax = plt.subplots(1)
ax.plot(SS, color = 'black', linewidth = 4)

yticks = ['','Wake','REM','N1','N2','N3']
ax.set_yticklabels(yticks)
ax.invert_yaxis()
plt.ylabel('stages', fontsize=32)
plt.yticks(fontsize=25)

plt.savefig("/home/b1044271/Columbia/Plots/Whole_Night_Stages.svg", dpi = 1200)
```
