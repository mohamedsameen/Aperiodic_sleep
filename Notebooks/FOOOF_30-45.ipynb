```python
# Import plot functions
from spiketools.plts.spikes import plot_isis
from spiketools.plts.trials import plot_rasters
from fooof.plts import plot_spectra
# Import the FOOOF object
from fooof import FOOOF
from fooof import FOOOFGroup

# Import a utility to download and load example data
from fooof.utils.download import load_fooof_data

import mne
import numpy as np
import os
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from spiketools.plts.utils import make_axes

import warnings
warnings.filterwarnings("ignore")
```

```python
# folder of the results
path_files = Path('/home/b1044271/Columbia/Results/PSDs/Cz/30-45/')

# Get names of all subject folder
sbj_folders = list(path_files.iterdir())
```

```python
def check_path(path):
    
    if not os.path.exists(path):
        os.mkdir(path)
```

```python
SETTINGS_F1={'max_n_peaks':8, 'aperiodic_mode':'fixed'}
SETTINGS_F2={'max_n_peaks':8, 'aperiodic_mode':'knee'}
SETTINGS_F3={'max_n_peaks':0, 'aperiodic_mode':'fixed'}
SETTINGS_F4={'max_n_peaks':0, 'aperiodic_mode':'knee'}

path_results = Path('/home/b1044271/Columbia/Results/FOOOF_interp/30-45')

fm1 = FOOOF(**SETTINGS_F1)
fm2 = FOOOF(**SETTINGS_F2)
fm3 = FOOOF(**SETTINGS_F3)
fm4 = FOOOF(**SETTINGS_F4)

fooof_a = {}
fooof_b = {}
fooof_c = {}
fooof_d = {}
```

```python
# get frequencies for all dimensions (x axis).
path_freqs = Path('/home/b1044271/Columbia/Results/PSDs/Freqs/30-45/')
f_files = list(path_freqs.iterdir())
f_files = sorted(f_files)
```

```python
N1_psd={}
N2_psd={}
N3_psd={}
AW_psd={}
RM_psd={}

T_PARAMS = ['T2', 'T5', 'T10', 'T15']
Stages   = ['N1','N2','N3','AW','RM']


for CUR_T_PARAM in T_PARAMS:
    for subj in sbj_folders:      
            
        S_path = str(subj)
        S_name = S_path[-4:]  
        check_path(path_results / S_name)
        for stage in Stages:
            check_path(path_results / S_name / stage )
            check_path(path_results / S_name / stage )
            check_path(path_results / S_name / stage /  'Knee')
            check_path(path_results / S_name / stage /  'NoKnee')

            freqs_d = np.load((path_freqs/(CUR_T_PARAM + '_30_45_freqres.npy')))
            
            A_psd = np.mean(np.load(path_files / subj / stage / (CUR_T_PARAM + '_30_45.npy')),0)

                
            fm1.fit(freqs_d, A_psd, [freqs_d[0] , freqs_d[-1]])
            fm1.save(S_name + '_'+ stage + '_'+ CUR_T_PARAM + '_'  +'30-45_NoKnee_8p',  file_path = path_results/ S_name/ stage/'NoKnee', save_results=True)

            fm2.fit(freqs_d, A_psd, [freqs_d[0] , freqs_d[-1]])
            fm2.save(S_name + '_'+  stage + '_'+ CUR_T_PARAM + '_' +'30-45_Knee_8p',  file_path = path_results/ S_name/stage/'Knee', save_results=True)

            fm3.fit(freqs_d, A_psd, [freqs_d[0] , freqs_d[-1]])
            fm3.save(S_name + '_'+ stage + '_'+ CUR_T_PARAM + '_'  +'30-45_NoKnee_0p',  file_path = path_results/ S_name/ stage/'NoKnee', save_results=True)

            fm4.fit(freqs_d, A_psd, [freqs_d[0] , freqs_d[-1]])
            fm4.save(S_name + '_'+  stage + '_'+ CUR_T_PARAM + '_' +'30-45_Knee_0p',  file_path = path_results/ S_name/stage/'Knee', save_results=True)

 
```
