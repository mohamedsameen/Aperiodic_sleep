```python
def check_path(path):
    
    if not os.path.exists(path):
        os.mkdir(path)
```

```python
# SET FOOOF PARAMETERS
SETTINGS_F1={'max_n_peaks':8, 'aperiodic_mode':'fixed'}
SETTINGS_F2={'max_n_peaks':8, 'aperiodic_mode':'knee'}
path_results = Path('/home/b1044271/Columbia/Results/FOOOF_interp')

fm1 = FOOOF(**SETTINGS_F1)
fm2 = FOOOF(**SETTINGS_F2, verbose = False)

fooof_a = {}
fooof_b = {}
```

```python
# get frequencies for all dimensions (x axis).
path_freqs = Path('/home/b1044271/Columbia/Results/PSDs/Freqs')
f_files = list(path_freqs.iterdir())
```

```python
N1_psd={}
N2_psd={}
N3_psd={}
AW_psd={}
RM_psd={}

F_PARAMS = ['F30', 'F45', 'F55', 'F70', 'F100']
T_PARAMS = ['T2', 'T5', 'T10', 'T15']
Stages   = ['N1','N2','N3','AW','RM']


for CUR_F_PARAM in F_PARAMS:
    for CUR_T_PARAM in T_PARAMS:
        for subj in sbj_folders:      
            
            S_path = str(subj)
            S_name = S_path[-4:]  
            check_path(path_results / S_name)
            for stage in Stages:
                check_path(path_results / S_name / stage)
                check_path(path_results / S_name / stage / 'Knee')
                check_path(path_results / S_name / stage / 'NoKnee')

                freqs_d = np.load((path_freqs/(CUR_T_PARAM + CUR_F_PARAM + '_freqres.npy')))
            
                A_psd = np.mean(np.load(path_files / subj / stage / (CUR_T_PARAM + '_' + CUR_F_PARAM + '.npy')),0)

                if freqs_d[-1] > 50:
                    freqs_d, A_psd = interpolate_spectrum(freqs_d, A_psd, [48, 52])
                
                fm1.fit(freqs_d, A_psd, [freqs_d[0] , freqs_d[-1]])
                fm1.save(S_name + '_'+ stage + '_'+ CUR_T_PARAM + '_' + CUR_F_PARAM +'_NoKnee',  file_path = path_results/ S_name/ stage/'NoKnee', save_results=True)

                fm2.fit(freqs_d, A_psd, [freqs_d[0] , freqs_d[-1]])
                fm2.save(S_name + '_'+  stage + '_'+ CUR_T_PARAM + '_' + CUR_F_PARAM +'_Knee',  file_path = path_results/ S_name/stage/'Knee', save_results=True)

 
```
