```python
# Import the FOOOF object
from fooof import FOOOF
import numpy as np

# Import a utility to download and load example data
from fooof.utils.download import load_fooof_data
import matplotlib.pyplot as plt

# Import required code for visualizing example models
from fooof import FOOOF
from fooof.sim.gen import gen_power_spectrum
from fooof.sim.utils import set_random_seed
from fooof.plts.spectra import plot_spectrum
from fooof.plts.annotate import plot_annotated_model
import seaborn
import random
import fooof.utils.params
import warnings
warnings.filterwarnings("ignore")
```

# 1. Simulate a PSD with a knee and then fit different models

```python
# Set random seed, for consistency generating simulated data
set_random_seed(21)

# Simulate example power spectra
freqs1, powers1 = gen_power_spectrum([1, 70], [1, 25, 1.25],
                                     [[10, 0.15, 1.], [30, 0.1, 2]])

# Plot one of the example power spectra
plot_spectrum(freqs1, powers1, log_powers=True,log_freqs =True, color='black')
plt.xlim(0.,1.9)
```

```python
fooof.utils.params.compute_knee_frequency(25, 1.25)
```

```python
# Model 1 - Knee Model
# Initialize a FOOOF object
fm1 = FOOOF(min_peak_height=0.05, aperiodic_mode='knee', verbose=False)
fm1.report(freqs1, powers1)
fm1.get_params('error')
fm1.get_params('r_squared')
fm1.get_params('aperiodic_params','exponent')

# Plot an annotated version of the power spectrum model
plot_annotated_model(fm1, plt_log=True)
plt.savefig('/home/b1044271/Figure_Modelsimulation-knee.svg')

plot_annotated_model(fm1,plt_log=True)
plt.savefig('/home/b1044271/Columbia/Plots/NOV23/Figure_Modelsimulation-knee_LogLog.svg')
```

```python
# Model 2 - Fixed Model
fm2 = FOOOF(min_peak_height=0.05, aperiodic_mode='fixed', verbose=False)
fm2.report(freqs1, powers1)
fm2.get_params('error')
fm2.get_params('r_squared')

# Plot an annotated version of the power spectrum model
plot_annotated_model(fm2,plt_log=True)
plt.savefig('/home/b1044271/Columbia/Plots/NOV23/Figure_Modelsimulation-Fixed_LogLog.svg')
```

# 2. Simulate several power spectra 

```python
# Set random seed, for consistency generating simulated data
set_random_seed(21)

f = np.zeros([100,139])* np.nan
p = np.zeros([100,139])* np.nan
Exp = np.zeros(100)* np.nan
Err = np.zeros(100)* np.nan
R2 = np.zeros(100)* np.nan

f2 = np.zeros([100,139])* np.nan
p2 = np.zeros([100,139])* np.nan
Exp2 = np.zeros(100)* np.nan
Err2 = np.zeros(100)* np.nan
R22 = np.zeros(100)* np.nan

f3 = np.zeros([100,139])* np.nan
p3 = np.zeros([100,139])* np.nan

Exp3 = np.zeros(100)* np.nan
Err3 = np.zeros(100)* np.nan
R23 = np.zeros(100)* np.nan

SETTINGS_F1 = {'max_n_peaks':8, 'aperiodic_mode':'knee'}

for s in np.arange(100):
    
    # SET PARAMETERS    
    c = random.sample(range(30), 2) # for ceonter freqs
    h = np.random.rand(2) # for height
    o = random.randint(-3, 3) # for offset
    e = random.uniform(1,5) # for exponent
    k = random.randint(100, 2000)# for Knee
    # for width
    w = []
    for i in range (0,2):
        x = round(random.uniform(0, 2), 2)
        w.append(x)
    

    # Simulate example power spectra
    f[s,0:139], p[s,0:139] = gen_power_spectrum([1, 70], [o, k, e], [[c[0], h[0], w[0]], [c[1], h[1], w[1]]])

    # Plot one of the example power spectra
    #plot_spectrum(freqs1, powers1, log_powers=True,log_freqs =True, color='black')

    # Initialize a FOOOF object
    fm = FOOOF(min_peak_height=0.05, aperiodic_mode='knee', verbose=False)
    fm.fit(f[s,0:139], p[s,0:139])
    # Report: fit the model, print the resulting parameters, and plot the reconstruction
    #fm.report(freqs, powers)
    Err[s] = fm.get_params('error')
    R2[s] = fm.get_params('r_squared')
    Exp[s] = fm.get_params('aperiodic_params','exponent')

    fm2 = FOOOF(min_peak_height=0.05, aperiodic_mode='fixed', verbose=False)
    fm2.fit(f[s,0:139], p[s,0:139])
    # Report: fit the model, print the resulting parameters, and plot the reconstruction
    #fm.report(freqs, powers)
    Err2[s] = fm2.get_params('error')
    R22[s] = fm2.get_params('r_squared')
    Exp2[s] = fm2.get_params('aperiodic_params','exponent')
    
    fm3 = FOOOF(min_peak_height=0.05, aperiodic_mode='fixed', verbose=False)
    fm3.fit(f[s,58:89], p[s,58:89])
    # Report: fit the model, print the resulting parameters, and plot the reconstruction
    #fm.report(freqs, powers)
    Err3[s] = fm3.get_params('error')
    R23[s] = fm3.get_params('r_squared')
    Exp3[s] = fm3.get_params('aperiodic_params','exponent')



# Plot an example power spectrum, with a model fit
#fm1.plot(plot_peaks='shade', peak_kwargs={'color' : 'green'}, plt_log=True)
#plt.savefig('/home/b1044271/Columbia/Plots/Knee_fit.svg')
#plt.savefig('/home/b1044271/Columbia/Plots/Orig_spctrm.svg')
```

## Plot R2, Error and Exponents

```python
array_list2 = [Exp3,Exp2,Exp]
array_list3 = [R23,R22,R2]
array_list5 = [Err3,Err2,Err]

titles =  ['simple','fixed', 'knee']

fig = plt.figure(figsize=(35, 10))  

ax = fig.add_subplot(1, 3, 1)
ax = seaborn.swarmplot(data=array_list3, size = 10)
ax.set_xticklabels(titles)
seaborn.scatterplot(x=[0,1,2], y=np.nanmean(array_list3,1), marker='X', color='black', s=100, zorder=4, legend=False);
plt.plot([array_list3[0], array_list3[1], array_list3[2]], color = 'grey', linewidth = 0.5);
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.ylim([0., 1.1])
plt.ylabel('R2', fontsize=24)


ax = fig.add_subplot(1, 3, 2)
ax = seaborn.swarmplot(data=array_list5, size = 10)
ax.set_xticklabels(titles)
seaborn.scatterplot(x=[0,1,2], y=np.nanmean(array_list5,1), marker='X', color='black', s=100, zorder=4, legend=False);
plt.plot([array_list5[0], array_list5[1], array_list5[2]], color = 'grey', linewidth = 0.5);
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.ylabel('Error', fontsize=24)
#plt.ylim([0.4, 3.5])


ax = fig.add_subplot(1, 3, 3)
ax = seaborn.swarmplot(data=array_list2, size = 10)
ax.set_xticklabels(titles)
seaborn.scatterplot(x=[0,1,2], y=np.nanmean(array_list2,1), marker='X', color='black', s=100, zorder=4, legend=False);
plt.plot([array_list2[0], array_list2[1], array_list2[2]], color = 'grey', linewidth = 0.5);
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.ylabel('exponent (a.u.)', fontsize=24)
#plt.ylim([0.4, 3.5])
```
